### Calculating the Information Value (IV):

```python
y = df['Default']
x = df.drop('Default', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.30,
                                                    random_state=42) 
```

```python
df_train = X_train.copy()
df_train['Default'] = y_train
```

```python
df_quant = df_train.loc[:,((df_train.dtypes == 'int32') |  
                        (df_train.dtypes == 'int64') | 
                        (df_train.dtypes == 'float64'))]

vars = df_quant.drop('Default',axis = 1).columns  

for var in vars:               
    name_var = 'fx_' + var 
    df_quant[name_var] = pd.qcut(df_quant[var], q=5, precision=2, duplicates='drop')
    df_quant = df_quant.drop(var, axis = 1)
    df_quant = df_quant.rename(columns={name_var: var})    
    
df_str = df_train.loc[:,((df_train.dtypes == 'object') | 
                    (df_train.dtypes == 'category') |
                    (df_train.dtypes == 'bool'))]

df_ = pd.concat([df_quant,df_str],axis = 1)
```

```python
to_drop = ['Default', 'LoanID']
vars = df_.drop(to_drop, axis = 1).columns  

List_IV = []
Feat_name = []

for var in vars: 
    
    df_aux = df_.copy() 
    df_aux['target'] = df['Default']
    df2 = df_aux.pivot_table(values='target',
                            index=var,
                            columns='Default',
                            aggfunc='count')
        
    df2 = df2.rename(columns={0:'#Target_0',
                            1:'#Target_1'})
    df2.fillna(0, inplace=True)

    df2['%Target_1'] = df2['#Target_1'] / df2['#Target_1'].sum()
    df2['%Target_0'] = df2['#Target_0'] / df2['#Target_0'].sum()
    df2['%Target_0'] = df2['%Target_0'].apply(lambda x: 0.01 if x == 0 else x) 
    df2['Odds'] = (df2['%Target_1'] / df2['%Target_0'])
    df2['Odds'] = df2.Odds.apply(lambda x: 0.01 if x == 0 else x)
    df2['LN(Odds)'] = np.log(df2['Odds'])
    df2['IV'] = ((df2['%Target_1'] - df2['%Target_0']) * df2['LN(Odds)'])
    df2['IV'] = np.where(df2['Odds'] == 0.01, 0 , df2['IV']) 
    Sum_IV = df2['IV'].sum().round(decimals=3)
    
    Feat_name.append(var)
    List_IV.append(Sum_IV)  
    
df_IV = pd.DataFrame(list(zip(Feat_name, List_IV)),
                     columns=['Feature', 'Sum_IV']).sort_values(by=['Sum_IV'],ascending=False)
df_IV = df_IV.reset_index( )
df_IV = df_IV.drop('index',axis=1)

def impact_IV (value):
    if value['Sum_IV'] < 0.02: return 'very weak'
    if value['Sum_IV'] < 0.1: return 'weak'
    if value['Sum_IV'] < 0.3: return 'medium'
    if value['Sum_IV'] < 0.5: return 'strong'   
    else: return 'very strong'

df_IV['Impact'] = df_IV.apply(impact_IV, axis=1)
df_IV
```

### Z test (hypothesis test):

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
def hypothesis_test (feature):
    
    df3 = df_.copy() 
    df3['target'] = df['Default']
    df4 = df3.pivot_table(values='target',
                        index= feature,
                        columns='Default',
                        aggfunc='count')

    df4 = df4.rename(columns={0:'#Target_0', 1:'#Target_1'})

    df4['Total'] = (df4['#Target_0'] + df4['#Target_1'])
        
    stat_list = []
    pval_list = []
    
    for i, index_str in enumerate(df4.index):
        
        nobs = np.array(df4['Total'][i:i+2])
        count = np.array(df4['#Target_1'][i:i+2])
        stat, pval = proportions_ztest(count, nobs)
      
        stat_list.append(stat)
        pval_list.append(pval)
        
        if i == df4.shape[0] - 2:
            break
            
    df_hyp = pd.DataFrame(list(zip(stat_list, pval_list)),columns=['Test_Z_stat','p-value'])

    return df_hyp
```

### Preparing the data for training:

```python
y = df['Default']

x_feat_name = ['Age','InterestRate','Income','MonthsEmployed','LoanAmount']

x = df[x_feat_name]
```

```python
def division (x, y):

    x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                        test_size= 0.30,
                                                        random_state= 42) 
    
    x_train, x_valid, y_train, y_valid = train_test_split(x, y, 
                                                        test_size= 0.20,
                                                        random_state= 42)    
    
    return x_train, x_test, y_train, y_test, x_valid, y_valid
```

```python
x_train, x_test, y_train, y_test, x_valid, y_valid = division (x, y)
```

```python
def sampling (x_train, y_train):
    
    under_tomek = TomekLinks(sampling_strategy='majority',n_jobs=4)

    x_train_tomek, y_train_tomek = under_tomek.fit_resample(x_train, y_train)

    rnd_under = RandomUnderSampler(sampling_strategy=0.35,
                                        random_state=42)
    x_train_under, y_train_under = rnd_under.fit_resample(x_train_tomek, y_train_tomek)

    smote = SMOTE(sampling_strategy= 1, random_state=42)

    x_train_over, y_train_over = smote.fit_resample(x_train_under, y_train_under)
    
    return x_train_over, y_train_over
```

```python
x_train_over, y_train_over = sampling (x_train, y_train)
```

### Gradient Boosting Classifier:

```python
def train_oversm_boosting (estimators, lean_rate, sample, max_dp): 
    

    model_gbc = GradientBoostingClassifier(n_estimators = estimators, 
                                           learning_rate = lean_rate,
                                           min_samples_leaf = 100,
                                           subsample = sample,           
                                           max_depth = max_dp,
                                           validation_fraction=0.3,
                                           random_state = 42)
                                            
    return model_gbc
```

```python
est = np.arange(50, 110, 10) 

leanrt = np.arange(0.30, 1.0, 0.10).round(decimals=2) 

samp = np.arange(0.30, 1.0, 0.20).round(decimals=2) 
 
maxdp = np.arange(3, 20, 1) 

# Defining the dictionary of hyperparameters: 
param_grid = dict(estimators=est,
                  lean_rate=leanrt,
                  sample =samp,
                  max_dp = maxdp)
```

```python
bo_space = {'estimators': hp.choice('estimators', param_grid['estimators']),
            'lean_rate': hp.choice('lean_rate', param_grid['lean_rate']),
            'sample': hp.choice('sample', param_grid['sample']),
            'max_dp': hp.choice('max_dp', param_grid['max_dp'])}
```

```python
def bo_gpc(param, x_train, y_train, x_valid, y_valid):   

    print ('tested_hyperparameters: ', param)
    model_bo= train_oversm_boosting(estimators=param['estimators'], 
                                    lean_rate=param['lean_rate'],
                                    sample =param['sample'],
                                    max_dp=param['max_dp'])   
    
    model_bo.fit(x_train, y_train)    
      
    ypred_train = model_bo.predict(x_train)
    
    ypred_valid = model_bo.predict(x_valid)
    
    Prec_train = precision_score(y_train, ypred_train)    
    Prec_valid = precision_score(y_valid, ypred_valid)    
    Prec_Diference = (Prec_train - Prec_valid)/Prec_train
        
    Recl_train = recall_score(y_train, ypred_train)    
    Recl_valid = recall_score(y_valid, ypred_valid)    
    Recl_Diference = (Recl_train - Recl_valid)/Recl_train
    
    D_p = math.sqrt((Prec_valid - 1)**2 + (Prec_Diference - 0)**2)
    D_r = math.sqrt((Recl_valid - 1)**2 + (Recl_Diference - 0)**2)
    D_t = D_p + D_r
    
    print('Recl_valid :', Recl_valid, 'Recl_train :', Recl_train,
         'Prec_valid :', Prec_valid, 'Prec_train :',Prec_train)
    
    
    return {'loss': D_t,'status': STATUS_OK}      
```

```python
wrapper_fun = lambda param: bo_gpc(param, x_train_over, y_train_over, x_valid, y_valid)
```

```python
trials = Trials()                  
bo_search =fmin(fn= wrapper_fun,       
                 space=bo_space,  
                 algo=tpe.suggest, 
                 max_evals=20,     
                 trials=trials)
```

```python
bo_better_hp = space_eval(bo_space, bo_search)
print('Better hyperparameters: ', bo_better_hp)
```

```python
model_gbc = train_oversm_boosting (estimators=90, lean_rate=0.4, sample=0.3, max_dp=4) 

model_gbc.fit(x_train_over, y_train_over)
```

```python
def performances (model, x_train, y_train, x_test, y_test, threshold):  

    if isinstance(model, tf.keras.Model):
        y_pred_train = (model.predict(x_train) > threshold).astype("int32")
        y_proba_train = model.predict(x_train)[:,0]
        
    else:
        try:
            y_proba_train = model.predict_proba(x_train)[:,1]
            y_pred_train = np.where(y_proba_train > threshold, 1, 0)
            
        except Exception as e:
            print(f'Model not supported: {e}')
            return None
   
    acc_train = accuracy_score(y_train, y_pred_train).round(decimals=2)
    roc_train = roc_auc_score(y_train, y_proba_train).round(decimals=2)
    recl_train = recall_score(y_train, y_pred_train).round(decimals=2)

    df_train = x_train.copy()
    df_train['Default'] = y_train
    df_train['Prob(Y=1)'] = y_proba_train
    
    train_prob_y_real_0 = df_train.loc[df_train.Default==0,"Prob(Y=1)"]
    train_prob_y_real_1 = df_train.loc[df_train.Default==1,"Prob(Y=1)"]

    KS_train = (stats.ks_2samp(train_prob_y_real_0, train_prob_y_real_1).statistic).round(decimals=2)

    if isinstance(model, tf.keras.Model):
       
        y_pred_test = (model.predict(x_test) > threshold).astype("int32")
        y_proba_test = model.predict(x_test)[:,0]
    
    else:
        try:   
            y_proba_test = model.predict_proba(x_test)[:,1]
            y_pred_test = np.where(y_proba_test > threshold, 1, 0)
            
        except Exception as e:
            print(f'Model not supported: {e}')
            return None
            
    acc_test = accuracy_score(y_test, y_pred_test).round(decimals=2)
    roc_test = roc_auc_score(y_test, y_proba_test).round(decimals=2)
    recl_test = recall_score(y_test, y_pred_test).round(decimals=2)

    df_test = x_test.copy()
    df_test['Default'] = y_test
    df_test['Prob(Y=1)'] = y_proba_test
    
    test_prob_y_real_0 = df_test.loc[df_test.Default==0,"Prob(Y=1)"]
    test_prob_y_real_1 = df_test.loc[df_test.Default==1,"Prob(Y=1)"]   
    
    KS_test = (stats.ks_2samp(test_prob_y_real_0, test_prob_y_real_1).statistic).round(decimals=2)

                  
    df_performance = pd.DataFrame({'Train':[recl_train, roc_train, KS_train, acc_train],
                              'Test':[recl_test, roc_test, KS_test, acc_test]},
                            index=['Recall','AUROC','KS','Accuracy'])
    
    df_performance['Variation'] = round(df_performance['Test'] / df_performance['Train'] - 1, 2)

    return df_performance 
```

### Bagging of Logistic Regression:

```python
def train_bagg_lr (estimators, samples, features): 
    
    model_base = LogisticRegression(max_iter=1000)   

    num_estimators = estimators          
  

    model_bagg_lr = BaggingClassifier(estimator = model_base, 
                                       n_estimators = num_estimators, 
                                       max_samples= samples,    
                                       max_features= features,   
                                       random_state = 42) 
    return model_bagg_lr
```

```python
est = np.arange(50, 210, 50)

samp = np.arange(0.30, 1, 0.20).round(decimals=2) 

feat = np.arange(0.60, 1.10, 0.40).round(decimals=2) 

param_grid = dict(estimators=est,
                 samples = samp,
                 features = feat)
```

```python
def bo_bagg(param, x_train, y_train, x_valid, y_valid):   

    print ('tested_hyperparameters: ', param)
    model_bo = train_bagg_lr (estimators=param['estimators'],
                              samples=param['samples'], 
                              features=param['features'])
    
    model_bo.fit(x_train, y_train)    
      
    ypred_train = model_bo.predict(x_train)
    
    ypred_valid = model_bo.predict(x_valid)
    
    Prec_train = precision_score(y_train, ypred_train)    
    Prec_valid = precision_score(y_valid, ypred_valid)    
    Prec_Diference = (Prec_train - Prec_valid)/Prec_train
        
    Recl_train = recall_score(y_train, ypred_train)    
    Recl_valid = recall_score(y_valid, ypred_valid)    
    Recl_Diference = (Recl_train - Recl_valid)/Recl_train
    
    D_p = math.sqrt((Prec_valid - 1)**2 + (Prec_Diference - 0)**2)
    D_r = math.sqrt((Recl_valid - 1)**2 + (Recl_Diference - 0)**2)
    D_t = D_p + D_r
    
    print('Recl_valid :', Recl_valid, 'Recl_train :', Recl_train,
         'Prec_valid :', Prec_valid, 'Prec_train :',Prec_train)
    
    return {'loss': D_t,'status': STATUS_OK}    
```

### Artificial Neural Networks:

```python
x_scaler = StandardScaler()

x_scaler.fit(x_train_over)
x_train_sc = x_scaler.transform(x_train_over)   
x_valid_sc = x_scaler.transform(x_valid)
x_test_sc = x_scaler.transform(x_test)
```

```python
def model_ANN (num_layers=1, num_neurons=8, fn_activ_hl='tanh', drop = 0.1):
    
    model = keras.Sequential()
    
    # First hidden layer:
    model.add(layers.Dense(num_neurons, 
                           activation=str(fn_activ_hl),
                           kernel_regularizer=keras.regularizers.l1_l2(),
                           input_shape=(x_train_sc.shape[1],)))
    model.add(layers.Dropout(drop))
    
    # Add more hidden layers:
    if num_layers > 1:
        for i in np.arange(2, num_layers, 1):
            model.add(layers.Dense(num_neurons, 
                                   activation=str(fn_activ_hl),
                                   kernel_regularizer=keras.regularizers.l1_l2()))
            model.add(layers.Dropout(drop))
    
    # Output layer:
    model.add(layers.Dense(1, activation='sigmoid'))
     
   
    model.compile(loss='binary_crossentropy', 
                  optimizer=keras.optimizers.Adam(),
                  metrics= [tf.keras.metrics.Recall(name='recl')])
    return model
```

```python
ehp_layers = np.arange(2, 6, 1)

ehp_neurons = np.power(2, np.arange(3, 6, 1))

ehp_fnactiv_hl = ['relu','tanh','linear']

drop = np.arange(0, 0.15, 0.05).round(decimals=2)

param_grid = dict(num_layers=ehp_layers,
                  num_neurons=ehp_neurons,
                  fn_activ_hl=ehp_fnactiv_hl,
                  drop = drop)
```

```python
bo_space4 = {'num_layers': hp.choice('num_layers', param_grid['num_layers']),
            'num_neurons': hp.choice('num_neurons', param_grid['num_neurons']),
            'fn_activ_hl': hp.choice('fn_activ_hl', param_grid['fn_activ_hl']),
             'drop': hp.choice('drop', param_grid['drop'])}
```

```python
rna_es = keras.callbacks.EarlyStopping(monitor="val_recl",    
                                        min_delta=0.005,
                                        patience = 20,
                                        verbose=1,
                                        mode='max',
                                        restore_best_weights=True)
```

```python
def bo_naa (param, x_train, y_train, x_valid, y_valid):    

    print ('Hiperparametros testados: ', param)
    model = model_ANN (num_layers=param['num_layers'], 
                          num_neurons=param['num_neurons'],
                          fn_activ_hl=param['fn_activ_hl'],
                            drop = param['drop']) 

    model.fit(x_train, 
            y_train,
            epochs=100,        
            batch_size=32,       
            validation_split= 0.2,  
            callbacks=[rna_es],        
            verbose=0) 

    ypred_train = (model.predict(x_train) > 0.5).astype("int32") 
        
    ypred_valid = (model.predict(x_valid) > 0.5).astype("int32")
                                                                 
    Prec_train = precision_score(y_train, ypred_train)    
    Prec_valid = precision_score(y_valid, ypred_valid)    
    Prec_Diference = (Prec_train - Prec_valid)/Prec_train
        
    Recl_train = recall_score(y_train, ypred_train)    
    Recl_valid = recall_score(y_valid, ypred_valid)    
    Recl_Diference = (Recl_train - Recl_valid)/Recl_train
    
    D_p = math.sqrt((Prec_valid - 1)**2 + (Prec_Diference - 0)**2)
    D_r = math.sqrt((Recl_valid - 1)**2 + (Recl_Diference - 0)**2)
    D_t = D_p + D_r
    
    print('Recl_valid :', Recl_valid, 'Recl_train :', Recl_train,
         'Prec_valid :', Prec_valid, 'Prec_train :',Prec_train)
    
    return {'loss': D_t,'status': STATUS_OK}    
```

```python
wrapper_fun4 = lambda param: bo_naa (param, x_train_sc, y_train_over, x_valid_sc, y_valid)
```

```python
def resume (models_list, index_list):

    Recall_Test = []
    AUROC_Test = []
    KS_Test = []
    Accur_Test = []
    Recall_Var = []
    AUROC_Var = []
    KS_Var = []
    Accur_Var = []

    models = models_list
    index = index_list

    for i in np.arange(0, 4):
        for model in models:
        
            result_1 = model['Test'][i]
    
            result_2 = model['Variation'][i]

            if i == 0:
                Recall_Test.append(result_1)  
                Recall_Var.append(result_2)             
            if i == 1:
                AUROC_Test.append(result_1)  
                AUROC_Var.append(result_2) 
            if i == 2:
                KS_Test.append(result_1)  
                KS_Var.append(result_2)             
            if i == 3:
                Accur_Test.append(result_1)  
                Accur_Var.append(result_2)     
    
    resume = pd.DataFrame({'Recall_Test': Recall_Test,'Recall_Var': Recall_Var,
                       'AUROC_Test': AUROC_Test ,'AUROC_Var': AUROC_Var,
                       'KS_Test': KS_Test,'KS_Var': KS_Var,
                       'Accur_Test': Accur_Test,'Accur_Var': Accur_Var}, index =index) 
    
    return resume.sort_values(by=['Recall_Test','Recall_Var','AUROC_Test', 'AUROC_Var','KS_Test', 'KS_Var'],
                               ascending=[False, False, False, False,False, False])
```

```python
models_list = [per_gbc, per_bagg, per_rand, per_ann]
index_list = ['Gradient_Boost', 'Logist_Regres', 'Random_For','Art_Neural_Net']

resume (models_list, index_list)
```

### Preparing the data for obtaining the different "score" variables (“featuring engineering”):

```python
def feat_qualit (df_train, feature, target, qcut):
    
    df_aux = df_train.copy() 
    
    name_feat = 'fx_' + feature  
    
    df_aux[name_feat] = pd.qcut(df_aux[feature], q= qcut, precision=2, duplicates='drop') 
    
    df_aux = df_aux.drop(feature, axis = 1)
    
    df_aux['target2'] = df_aux[target]
    
    df2 = df_aux.pivot_table(values='target2',
                            index= name_feat,
                            columns= target,
                            aggfunc='count')
    
    df2 = df2.rename(columns={0:'Target_0', 1:'Target_1'})
    df2['Total'] = (df2['Target_0'] + df2['Target_1'])
    df2['% Rate_Target_1'] = (df2['Target_1'] / df2['Total']).round(decimals=2)
    df2['Group'] = np.arange(1, qcut+1, 1) 
    
    return df2    
```
